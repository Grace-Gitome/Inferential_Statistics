## Skewness
Skewness is a statistic that measures the asymmetry of a distribution. A distribution is asymmetrical when its left and right side are not mirror images.

A distribution can have **right (or positive)**, **left (or negative)**, or **zero skewness**.

#### What is zero skew?
When a distribution has zero skew, it is symmetrical. Its left and right sides are mirror images. The easiest way to check if a variable has a skewed distribution is to plot it in a histogram.

The distribution is approximately symmetrical, with the observations distributed similarly on the left and right sides of its peak. Therefore, the distribution has approximately zero skew.

**NOTE**: In a distribution with zero skew, the mean and median are equal. (Mean = Median)

#### What is right skew (positive skew)?
A right-skewed distribution is longer on the right side of its peak than on its left. It indicates that there are observations at one of the extreme ends of the distribution, but that they’re relatively infrequent. A right-skewed distribution has a long tail on its right side.

**NOTE**: In a distribution with a positive skew, the mean is is almost always greater than its median. (Mean > Median)
that is because extreme values (values in the tail) affect the mean more than the median.

#### What is left skew (left skew)?
A left-skewed distribution is longer on the left side of its peak than on its right.The mean of a left-skewed distribution is almost always less than its median. (Mean < Median>)

#### How to calculate skewness
One of the simplest is Pearson’s median skewness. It takes advantage of the fact that the mean and median are unequal in a skewed distribution.

#### What to do if your data is skewed
One reason you might check if a distribution is skewed is to verify whether your data is appropriate for a certain statistical procedure.

There are three choices if the statistical procedure requires a normal distribution and your data is skewed:

1. **Do nothing**

Many statistical tests, including t tests, ANOVAs, and linear regressions, aren’t very sensitive to skewed data. Especially if the skew is mild or moderate, it may be best to ignore it.

2. **Use a different model**

Choose a model that doesn’t assume a normal distribution. Non-parametric tests or generalized linear models could be more appropriate for your data.

3. **Transform the variable**

Another option is to transform a skewed variable so that it’s less skewed. “Transform” means to apply the same function to all the observations of a variable.
